```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('miles.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (12,4))
plt.plot(df.miles)
```

```python
sns.histplot(df.miles, kde = True)
```

```python
polpulation_mean = df.miles.mean()
polpulation_std_dev = df.miles.std()
polpulation_mean, polpulation_std_dev
```

```python
sample_size = 50

sample = df.sample(sample_size, random_state = 100)
sample.head()
```

```python
sample_mean = sample.miles.mean()
sample_std_dev = sample.miles.std()
sample_mean, sample_std_dev
```

```python
import scipy.stats as st
```

```python
confidence_z_score_95 = st.norm.ppf(0.975)
confidence_z_score_95
```

```python
error_margin = confidence_z_score_95 * (polpulation_std_dev/np.sqrt(sample_size))
error_margin
```

```python
upper_conf_int = sample_mean + error_margin
lower_conf_int = sample_mean - error_margin

lower_conf_int, upper_conf_int
```
