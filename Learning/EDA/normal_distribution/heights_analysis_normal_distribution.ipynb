```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heights.csv")
df.head()
```

```python
df.shape
```

```python
mean = round(df.height.mean() ,2)
std = round(df.height.std() ,2)
mean, std
```

```python
sns.histplot(df.height, kde = True)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
plt.axvline(mean + std, color='green', linestyle='--', linewidth=1)
plt.axvline(mean - std, color='green', linestyle='--', linewidth=1)
plt.axvline(mean + 2*std, color='blue', linestyle='--', linewidth=1)
plt.axvline(mean - 2*std, color='blue', linestyle='--', linewidth=1)
plt.axvline(mean + 3*std, color='orange', linestyle='--', linewidth=1)
plt.axvline(mean - 3*std, color='orange', linestyle='--', linewidth=1)
plt.show()
```

```python
GL = round(mean-std, 2)
GR = round(mean+std, 2)
GL,GR
```

```python
BL = round(mean-2*std, 2)
BR = round(mean+2*std, 2)
BL, BR
```

```python
OL = round(mean -3*std, 2)
OR = round(mean+3*std, 2)
OL, OR
```

```python
green = df[(df.height>GL) & (df.height<GR)]
green.shape
```

```python
blue = df[(df.height>BL) & (df.height<BR)]
blue.shape
```

```python
orange = df[(df.height>OL) & (df.height<OR)]
orange.shape
```
