```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heights_with_outliers.csv')
df.head(5)
```

```python
mean = round(df.height.mean(),2)
std = round(df.height.std(),2)

mean, std
```

```python
sns.histplot(df.height, kde = True)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
plt.axvline(mean + std, color='green', linestyle='--', linewidth=1)
plt.axvline(mean - std, color='green', linestyle='--', linewidth=1)
plt.axvline(mean + 2*std, color='blue', linestyle='--', linewidth=1)
plt.axvline(mean - 2*std, color='blue', linestyle='--', linewidth=1)
plt.axvline(mean + 3*std, color='orange', linestyle='--', linewidth=1)
plt.axvline(mean - 3*std, color='orange', linestyle='--', linewidth=1)
plt.show()
```

```python
lower = round(mean - 3*std, 2)
upper = round(mean + 3*std,2)
lower, upper
```

```python
outliers = df[(df.height<lower) | (df.height>upper)]
outliers
```

```python
outliers.shape
```

```python
normalized = df[(df.height>lower) & (df.height<upper)]
normalized.shape
```

```python
sns.histplot(normalized, kde = True)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
plt.axvline(mean + std, color='green', linestyle='--', linewidth=1)
plt.axvline(mean - std, color='green', linestyle='--', linewidth=1)
plt.axvline(mean + 2*std, color='blue', linestyle='--', linewidth=1)
plt.axvline(mean - 2*std, color='blue', linestyle='--', linewidth=1)
plt.axvline(mean + 3*std, color='orange', linestyle='--', linewidth=1)
plt.axvline(mean - 3*std, color='orange', linestyle='--', linewidth=1)
plt.show()
```
