```python
import pandas as pd

df = pd.read_csv('heights.csv')
df.head()
```

```python
Q1, Q3 = df.height.quantile([0.25, 0.75])
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3+1.5*IQR

lower_limit, upper_limit
```

```python
outliers = df[(df.height<lower_limit) | (df.height>upper_limit)]
outliers
```
