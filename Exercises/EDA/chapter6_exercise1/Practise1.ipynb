### STD on age

```python
left1 = round(meana-stda, 2)
right1 = round(meana+stda, 2)
print("the age range withhin 1 std of the mean: ", left1, right1)
```

```python
left2 = round(meana-2*stda, 2)
right2 = round(meana+2*stda, 2)
print("the age range withhin 2 std of the mean: ", left2, right2)
```

```python
left3 = round(meana-3*stda, 2)
right3 = round(meana+3*stda, 2)
print("the age range withhin 3 std of the mean: ", left3, right3)
```

### STD on player_rating

```python
mean_pr = round(df['player_rating'].mean(), 1)
std_pr = round(df['player_rating'].std(), 1)
```

```python
left11 = round(mean_pr-std_pr, 2)
right11 = round(mean_pr+std_pr, 2)
print("the age range withhin 1 std of the mean: ", left11, right11)

left22 = round(mean_pr-2*std_pr, 2)
right22 = round(mean_pr+2*std_pr, 2)
print("the age range withhin 2 std of the mean: ", left22, right22)

left33 = round(mean_pr-3*std_pr, 2)
right33 = round(mean_pr+3*std_pr, 2)
print("the age range withhin 3 std of the mean: ", left33, right33)
```

### SHOW OUTLIERS

```python
age_outliers = df[(df.age<left3) | (df.age>right3)]
```

```python
player_rating_outliers = df[(df.player_rating<left33) | (df.player_rating>right33)]
```

### REMOVE OUTLIERS

```python
df = df[(df.age>left3) & (df.age<right3)]
```

```python
df = df[(df.player_rating>left33) & (df.player_rating<right33)]
```
